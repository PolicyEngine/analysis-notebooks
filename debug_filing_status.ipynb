```python
from policyengine_us import Microsimulation
import pandas as pd
```

```python
baseline = Microsimulation(dataset="hf://policyengine/policyengine-us-data/enhanced_cps_2024.h5")
```

```python
variables = ["household_id", "household_size", "age", "age_head", "age_spouse", "is_married", "tax_unit_dependents", "filing_status", "is_tax_unit_spouse", "tax_unit_married", "is_tax_unit_head", "household_weight", "person_household_id", "person_tax_unit_id", "tax_unit_weight", "tax_unit_id", "tax_unit_household_id"]
             
```

```python
hh_variables = ["household_id", "household_size", "age", "age_head", "age_spouse", "is_married", "tax_unit_dependents", "is_tax_unit_spouse", "tax_unit_married", "is_tax_unit_head", "household_weight", "person_household_id", "person_tax_unit_id", "tax_unit_weight", "tax_unit_id", "tax_unit_household_id"]
```

```python
person_df = baseline.calculate_dataframe(variables, map_to="person", period=2026)
person_df = person_df.sort_values('household_id')
tax_unit_df = baseline.calculate_dataframe(variables, map_to="tax_unit",  period=2026)
tax_unit_df = tax_unit_df.sort_values('household_id')
spm_unit_df = baseline.calculate_dataframe(hh_variables, map_to="spm_unit", period=2026)
spm_unit_df = tax_unit_df.sort_values('household_id')
household_df = baseline.calculate_dataframe(hh_variables, map_to="household", period=2026)
household_df = household_df.sort_values('household_id')
family_df = baseline.calculate_dataframe(hh_variables, map_to="family", period=2026)
family_df = family_df.sort_values('household_id')

```
