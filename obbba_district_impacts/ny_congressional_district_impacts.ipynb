```python
# Create NY congressional district map visualization
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

# Load the hexagon shapefile
hex_gdf = gpd.read_file('HexCDv31/HexCDv31.shp')

# Convert GEOID to integer for merging
hex_gdf['cd_geoid'] = hex_gdf['GEOID'].astype(int)

# Filter to only NY districts (state FIPS code 36)
ny_hex = hex_gdf[hex_gdf['GEOID'].str.startswith('36')].copy()

print(f"Found {len(ny_hex)} NY congressional districts in shapefile")

# Prepare district impacts for merging
district_impacts_merge = district_impacts.reset_index()
district_impacts_merge['cd_geoid'] = district_impacts_merge['congressional_district_geoid'].astype(int)

# Merge the geographic data with impact data
ny_map = ny_hex.merge(
    district_impacts_merge[['cd_geoid', 'avg_income_change', 'sample_size']], 
    on='cd_geoid', 
    how='left'
)

print(f"\nMerged data: {len(ny_map)} districts")
print(f"Districts with impact data: {ny_map['avg_income_change'].notna().sum()}")
```

```python
# Plot NY congressional district income changes
fig, ax = plt.subplots(figsize=(12, 10))

# Create the choropleth map
ny_map.plot(
    column='avg_income_change',
    cmap='RdYlGn',  # Red for negative (income loss), Green for positive (income gain)
    legend=True,
    ax=ax,
    edgecolor='black',
    linewidth=0.5,
    missing_kwds={'color': 'lightgray'}
)

# Add district labels
for idx, row in ny_map.iterrows():
    # Get the centroid of each district
    centroid = row.geometry.centroid
    # Extract district number from GEOID (last 2 digits)
    district_num = str(row['GEOID'])[-2:]
    ax.text(centroid.x, centroid.y, district_num, 
            fontsize=8, ha='center', va='center', weight='bold')

ax.set_title('Average Change in Net Income by NY Congressional District\n(Reform vs Current Law)', 
             fontsize=14, weight='bold')
ax.axis('off')

plt.tight_layout()
plt.savefig('ny_district_income_impact.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\nMap saved as 'ny_district_income_impact.png'")
```
