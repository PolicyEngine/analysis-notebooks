```python
def calculate_poverty_metrics(baseline_sim, reform_sim, period=2026):
    """Calculate poverty metrics for a given reform compared to baseline"""
    baseline_poverty = baseline_sim.calculate("in_poverty", map_to="person", period=period)
    reformed_poverty = reform_sim.calculate("in_poverty", map_to="person", period=period)
    is_child = baseline_sim.calculate("is_child", map_to="person", period=period)
    
    # Baseline poverty rates
    baseline_poverty_rate = baseline_poverty.mean()
    baseline_child_poverty_rate = baseline_poverty[is_child].mean()
    
    # Reformed poverty rates
    reformed_poverty_rate = reformed_poverty.mean()
    reformed_child_poverty_rate = reformed_poverty[is_child].mean()
    
    # Overall poverty change
    poverty_change = reformed_poverty_rate - baseline_poverty_rate
    
    # Child poverty change
    child_poverty_change = reformed_child_poverty_rate - baseline_child_poverty_rate
    
    return {
        'baseline_poverty_rate': baseline_poverty_rate,
        'baseline_child_poverty_rate': baseline_child_poverty_rate,
        'poverty_change': poverty_change,
        'child_poverty_change': child_poverty_change,
        'final_poverty_rate': reformed_poverty_rate,
        'final_child_poverty_rate': reformed_child_poverty_rate
    }
```

```python
def run_simulations(baseline_reform, reforms_dict, dataset=None):
    """Run simulations for baseline and reforms, return poverty metrics"""
    # Create simulations
    if dataset:
        baseline_sim = Microsimulation(reform=baseline_reform, dataset=dataset)
        reform_sims = {name: Microsimulation(reform=reform, dataset=dataset) 
                      for name, reform in reforms_dict.items()}
    else:
        baseline_sim = Microsimulation(reform=baseline_reform)
        reform_sims = {name: Microsimulation(reform=reform) 
                      for name, reform in reforms_dict.items()}
    
    # Calculate metrics for each reform
    results = {}
    baseline_rates = None
    
    for name, reform_sim in reform_sims.items():
        metrics = calculate_poverty_metrics(baseline_sim, reform_sim)
        
        # Store baseline rates from first calculation
        if baseline_rates is None:
            baseline_rates = {
                'baseline_poverty_rate': metrics['baseline_poverty_rate'],
                'baseline_child_poverty_rate': metrics['baseline_child_poverty_rate']
            }
        
        results[name] = metrics
    
    return results, baseline_rates
```
