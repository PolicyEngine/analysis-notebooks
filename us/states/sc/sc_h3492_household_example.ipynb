# SC H.3492 Partially Refundable EITC - Household Examples

This notebook demonstrates the impact of South Carolina H.3492 on individual households.

## Reform Summary
- **Baseline**: SC EITC = 125% of federal EITC (non-refundable, limited by tax liability)
- **Reform**: Makes 25% of the excess SC EITC (above tax liability) refundable

## Key Formula
- If SC EITC > tax liability:
  - Non-refundable portion = tax liability
  - Refundable portion = 25% × (SC EITC - tax liability)

## Bill Reference
[SC H.3492 (126th Session, 2025-2026)](https://www.scstatehouse.gov/sess126_2025-2026/bills/3492.htm)

```python
from policyengine_us import Simulation
from policyengine_us.reforms.states.sc.h3492.sc_h3492_eitc_refundable import sc_h3492_eitc_refundable
import pandas as pd
import numpy as np
```

```python
def create_reform():
    """Create the SC H.3492 reform."""
    return sc_h3492_eitc_refundable
```

## Example 1: Single Parent with One Child, $20,000 Income

This is the primary example from the issue specification. A low-income single parent should see significant benefit from the partial refundability.

```python
situation_1 = {
    "people": {
        "parent": {
            "age": {2025: 30},
            "employment_income": {2025: 20_000}
        },
        "child": {
            "age": {2025: 5},
            "is_tax_unit_dependent": {2025: True}
        }
    },
    "tax_units": {
        "tax_unit": {
            "members": ["parent", "child"]
        }
    },
    "families": {
        "family": {
            "members": ["parent", "child"]
        }
    },
    "spm_units": {
        "spm_unit": {
            "members": ["parent", "child"]
        }
    },
    "households": {
        "household": {
            "members": ["parent", "child"],
            "state_code": {2025: "SC"}
        }
    }
}

baseline_1 = Simulation(situation=situation_1)
reform_1 = Simulation(situation=situation_1, reform=create_reform())

print("="*70)
print("EXAMPLE 1: Single Parent with 1 child (age 5), $20k income")
print("="*70)
print(f"\nFederal EITC: ${baseline_1.calculate('eitc', 2025)[0]:,.2f}")
print(f"SC EITC (125% of federal): ${baseline_1.calculate('sc_eitc', 2025)[0]:,.2f}")
print(f"\nBaseline (non-refundable):")
print(f"  SC tax before non-refundable credits: ${baseline_1.calculate('sc_income_tax_before_non_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC non-refundable credits: ${baseline_1.calculate('sc_non_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC refundable credits: ${baseline_1.calculate('sc_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC income tax: ${baseline_1.calculate('sc_income_tax', 2025)[0]:,.2f}")
print(f"  Household net income: ${baseline_1.calculate('household_net_income', 2025)[0]:,.2f}")
print(f"\nReform (25% of excess refundable):")
print(f"  SC non-refundable credits: ${reform_1.calculate('sc_non_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC refundable credits: ${reform_1.calculate('sc_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC income tax: ${reform_1.calculate('sc_income_tax', 2025)[0]:,.2f}")
print(f"  Household net income: ${reform_1.calculate('household_net_income', 2025)[0]:,.2f}")
print(f"\n>>> BENEFIT FROM REFORM: ${reform_1.calculate('household_net_income', 2025)[0] - baseline_1.calculate('household_net_income', 2025)[0]:,.2f}")
```

## Example 5: Zero Tax Liability (Maximum Refund Scenario)

A household with zero tax liability represents the maximum refund scenario - the entire excess EITC is subject to the 25% refund rate.

```python
situation_5 = {
    "people": {
        "parent": {
            "age": {2025: 25},
            "employment_income": {2025: 10_000}
        },
        "child": {
            "age": {2025: 2},
            "is_tax_unit_dependent": {2025: True}
        }
    },
    "tax_units": {
        "tax_unit": {
            "members": ["parent", "child"]
        }
    },
    "families": {
        "family": {
            "members": ["parent", "child"]
        }
    },
    "spm_units": {
        "spm_unit": {
            "members": ["parent", "child"]
        }
    },
    "households": {
        "household": {
            "members": ["parent", "child"],
            "state_code": {2025: "SC"}
        }
    }
}

baseline_5 = Simulation(situation=situation_5)
reform_5 = Simulation(situation=situation_5, reform=create_reform())

print("="*70)
print("EXAMPLE 5: Single Parent with 1 child, $10k income (low/zero tax)")
print("="*70)
print(f"\nFederal EITC: ${baseline_5.calculate('eitc', 2025)[0]:,.2f}")
print(f"SC EITC (125% of federal): ${baseline_5.calculate('sc_eitc', 2025)[0]:,.2f}")
print(f"\nBaseline (non-refundable):")
print(f"  SC tax before non-refundable credits: ${baseline_5.calculate('sc_income_tax_before_non_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC non-refundable credits: ${baseline_5.calculate('sc_non_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC refundable credits: ${baseline_5.calculate('sc_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC income tax: ${baseline_5.calculate('sc_income_tax', 2025)[0]:,.2f}")
print(f"  Household net income: ${baseline_5.calculate('household_net_income', 2025)[0]:,.2f}")
print(f"\nReform (25% of excess refundable):")
print(f"  SC non-refundable credits: ${reform_5.calculate('sc_non_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC refundable credits: ${reform_5.calculate('sc_refundable_credits', 2025)[0]:,.2f}")
print(f"  SC income tax: ${reform_5.calculate('sc_income_tax', 2025)[0]:,.2f}")
print(f"  Household net income: ${reform_5.calculate('household_net_income', 2025)[0]:,.2f}")

sc_eitc = baseline_5.calculate('sc_eitc', 2025)[0]
tax_before = baseline_5.calculate('sc_income_tax_before_non_refundable_credits', 2025)[0]
expected_refund = 0.25 * max(0, sc_eitc - tax_before)
print(f"\n>>> BENEFIT FROM REFORM: ${reform_5.calculate('household_net_income', 2025)[0] - baseline_5.calculate('household_net_income', 2025)[0]:,.2f}")
print(f"\n*** Expected refund calculation: 25% × (${sc_eitc:,.2f} - ${tax_before:,.2f}) = ${expected_refund:,.2f} ***")
```

## Summary Table

```python
examples = [
    ("1: 1 child, $20k", baseline_1, reform_1, True),
    ("2: 1 child, $40k", baseline_2, reform_2, False),
    ("3: 2 children, $15k", baseline_3, reform_3, True),
    ("4: Childless, $12k", baseline_4, reform_4, True),
    ("5: 1 child, $10k", baseline_5, reform_5, True),
    ("6: 3 children, $17k", baseline_6, reform_6, True),
    ("7: 1 child, $0", baseline_7, reform_7, False),
]

summary_data = []
for desc, baseline, reform, expected_benefit in examples:
    baseline_income = baseline.calculate('household_net_income', 2025)[0]
    reform_income = reform.calculate('household_net_income', 2025)[0]
    benefit = reform_income - baseline_income
    federal_eitc = baseline.calculate('eitc', 2025)[0]
    sc_eitc = baseline.calculate('sc_eitc', 2025)[0]
    sc_tax = baseline.calculate('sc_income_tax_before_non_refundable_credits', 2025)[0]
    reform_refundable = reform.calculate('sc_refundable_credits', 2025)[0]
    actual_benefit = benefit > 0.5
    status = "PASS" if actual_benefit == expected_benefit else "FAIL"
    
    summary_data.append({
        "Example": desc,
        "Fed EITC": f"${federal_eitc:,.0f}",
        "SC EITC": f"${sc_eitc:,.0f}",
        "SC Tax": f"${sc_tax:,.0f}",
        "Refund": f"${reform_refundable:,.0f}",
        "Benefit": f"${benefit:,.2f}",
        "Status": status,
    })

df_summary = pd.DataFrame(summary_data)
print("="*100)
print("SUMMARY: SC H.3492 Partially Refundable EITC Household Examples")
print("="*100)
print(df_summary.to_string(index=False))
print("="*100)

all_pass = all(row["Status"] == "PASS" for row in summary_data)
print(f"\nAll tests passed: {all_pass}")

print("\n" + "-"*70)
print("KEY INSIGHT - Refund Formula:")
print("-"*70)
print("Refundable portion = 25% × max(0, SC EITC - SC tax liability)")
print("\nHouseholds benefit when their SC EITC exceeds their tax liability.")
print("The lower the tax liability relative to the SC EITC, the larger the refund.")
print("-"*70)
```
